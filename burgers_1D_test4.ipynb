```python
import numpy as np
import matplotlib.pyplot as plt
import math

# variables and discretization parameters
nt = 5000
nx = 101
c =(0.01/math.pi)
nu = 0.1
dt = 0.001

dx = 2 / (nx - 1)
x = np.linspace(-1, 1, nx)

u = np.zeros(nx)
un = np.zeros(nx)
uf = np.zeros((nt ,nx))


# assigning initial conditions
#u[int(0.75 / dx):int(1.25 / dx + 1)] = 5

u = -np.sin(math.pi * x)
uf[0,:]= u




# Burgers equation: ∂u/∂t + u∂u/∂x = nu∂²u/∂x²

for n in range(nt):
    un = u.copy()
    for i in range(1, nx - 1):
        u[i] = (un[i] - un[i] * dt / dx * (un[i] - un[i - 1])) + (nu * dt / (dx**2)) * (un[i + 1] - 2 * un[i] + un[i - 1])
        uf[n,i]=u[i]
    
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(x, u, label="u Solution")
plt.title("1D Burgers' Equation Solution")
plt.xlabel("X")
plt.ylabel("u")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# variables and discretization parameters
nt = 200
nx = 51
c = 0.1# (0.01/math.pi)
nu = 0.01
dt = 0.001

dx = 2 / (nx - 1)
x = np.linspace(0, 2, nx)

u = np.ones(nx)
un = np.zeros(nx)
uf = np.zeros((nt ,nx))

# assigning initial conditions
u[int(0.75 / dx):int(1.25 / dx + 1)] = 5
#u =  -np.sin(math.pi * x)
uf[0,:]= u

# Burgers equation: ∂u/∂t + u∂u/∂x = nu∂²u/∂x²

for n in range(nt):
    un = u.copy()
    for i in range(1, nx - 1):
        u[i] = (un[i] - un[i] * dt / dx * (un[i] - un[i - 1])) + (nu * dt / (dx**2)) * (un[i + 1] - 2 * un[i] + un[i - 1])
        uf[n,i]=u[i]
    u[0] = 1
    u[-1]= 1
    uf[n,0]= 1
    uf[n,-1]=1
    
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(x, u, label="u Solution")
plt.title("1D Burgers' Equation Solution")
plt.xlabel("X")
plt.ylabel("u")
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(5,3), dpi=3000)
sns.heatmap(uf.T, cmap='jet')
```
